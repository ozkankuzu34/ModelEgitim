```python
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization,GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
```

```python
# InceptionV3 tabanını yükle
base_model = InceptionV3(weights="imagenet", include_top=False, input_shape=(299, 299, 3))

# Base model'i değiştirmez (freeze) hale getir
base_model.trainable = False

# Özelleştirilmiş katmanlar
x = base_model.output
x = GlobalAveragePooling2D()(x)
#x = Dense(4096, activation='relu')(x)
x = Dense(1024, activation='relu')(x)
#x = BatchNormalization()(x)
x = Dropout(0.5)(x)
x = Dense(3, activation='softmax')(x)  # Binary classification için

model = Model(inputs=base_model.input, outputs=x)

# Modeli derle
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

# Model özeti
model.summary()
```

```python
# Veri augmentasyonu
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)

# Veri generatorleri
train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/dataset10/train',
    target_size=(299, 299),
    batch_size=16,
    class_mode='categorical'
)

validation_generator = val_datagen.flow_from_directory(
    '/content/drive/MyDrive/dataset10/validation',
    target_size=(299, 299),
    batch_size=16,
    class_mode='categorical'
)
```

```python
# Modeli eğit
history = model.fit(
    train_generator,

    validation_data=validation_generator,

    epochs=10
)

# Eğitilmiş modeli kaydet
#model.save("vgg16_peach_leaf_model11.h5")
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Test verilerini hazırlayın
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    '/content/drive/My Drive/dataset10/test',
    target_size=(299, 299),
    batch_size=1,
    class_mode='categorical',
    shuffle=False  # Tahmin ve gerçek sıralarını eşleştirmek için shuffle=False
)

# Modelin test verileri üzerindeki tahminlerini alın
predictions = model.predict(test_generator)

# Tahmin edilen sınıf indekslerini alın
predicted_classes = np.argmax(predictions, axis=1)

# Gerçek etiketleri alın
true_classes = test_generator.classes
class_labels = list(test_generator.class_indices.keys())

# Confusion Matrix oluşturun
conf_matrix = confusion_matrix(true_classes, predicted_classes)

# Confusion Matrix'i görselleştirin
plt.figure(figsize=(8, 8))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
plt.show()

# Sınıflandırma raporunu yazdırın
print(classification_report(true_classes, predicted_classes, target_names=class_labels))
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np

def preprocess_image(image_path, target_size=(299, 299)):
    img = load_img(image_path, target_size=target_size)
    x = img_to_array(img)
    x = x.astype('float32') / 255.
    x = np.expand_dims(x, axis=0)
    return x

x = preprocess_image('/content/drive/My Drive/Dataset/Test/Test/Healthy/6029_png.rf.ebf75ea2748f87fe2a8c81e5b094567d.jpg')
```
